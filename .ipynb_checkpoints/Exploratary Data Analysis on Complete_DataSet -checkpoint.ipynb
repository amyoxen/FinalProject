```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
store = pd.HDFStore('all_data.h5')
all_data = store['all_data']
```

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.boxplot(x='prepay', y='loan_amount', hue="platform_id", data=all_data)
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
plt.title("BoxPlot on loan amount over platforms")
plt.tight_layout()
```

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.boxplot(x='default', y='loan_amount', hue="platform_id", data=all_data)
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
plt.title("BoxPlot on loan amount over platforms")
plt.tight_layout()
```

```python
all_data_completed = all_data.drop(['Orig_Date', 'orig_date'], axis=1)
all_data_completed = all_data_completed.drop(['loan_status','report_date', 'bom','ppmt','ipmt','co_amt','prepay_amt','eom' ], axis=1)
all_data_completed = all_data_completed.dropna()
all_data_completed.head(10)
```
