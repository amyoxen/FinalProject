```python
import pandas as pd
import os
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_excel(os.path.join("data", "PROPRIETARY_INFORMATION_Orig_Holdout_Data.xlsx"),engine='openpyxl',sheet_name='Sheet1')
```

```python
li = []
df_a_orig_date = pd.read_csv('data/PROPRIETARY_INFORMATION_Orig_Holdout_Data_Plat_A_Orig_Date.csv', index_col=None, header=0)
df_a_orig_date['Orig_year'] = pd.DatetimeIndex(pd.to_datetime(df_a_orig_date['Orig_Date'].str.replace('-',''),format='%b%y')).year
df_a_orig_date['Orig_month'] = pd.DatetimeIndex(pd.to_datetime(df_a_orig_date['Orig_Date'].str.replace('-',''),format='%b%y')).month
df_a_orig_date.drop('Orig_Date', inplace=True, axis=1)
li.append(df_a_orig_date )


df_b_orig_date = pd.read_csv('data/PROPRIETARY_INFORMATION_Orig_Holdout_Data_Plat_B_Orig_Date.csv', index_col=None, header=0)
df_b_orig_date['Orig_year'] = pd.DatetimeIndex(pd.to_datetime(df_b_orig_date['orig_date'].str.replace('-',''),format='%Y%m')).year
df_b_orig_date['Orig_month'] = pd.DatetimeIndex(pd.to_datetime(df_b_orig_date['orig_date'].str.replace('-',''),format='%Y%m')).month
df_b_orig_date.drop('orig_date', inplace=True, axis=1)
li.append(df_b_orig_date)


df_c_orig_date = pd.read_csv('data/PROPRIETARY_INFORMATION_Orig_Holdout_Data_Plat_C_Orig_Date.csv', index_col=None, header=0)
df_c_orig_date['Orig_year'] = pd.DatetimeIndex(pd.to_datetime(df_c_orig_date['orig_date'],format='%Y%m')).year
df_c_orig_date['Orig_month'] = pd.DatetimeIndex(pd.to_datetime(df_c_orig_date['orig_date'],format='%Y%m')).month
df_c_orig_date.drop('orig_date', inplace=True, axis=1)
li.append(df_c_orig_date)

```

```python
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame = frame.merge(df, on='unique_id', how='left')
```

Treament of Data

```python
frame['platform_id'] = frame['unique_id'].astype(str).str[0]
```

```python
store = pd.HDFStore('df_all_orig_means.h5')
means = store['df_all_orig_means'] 
```

```python
frame['num_tot_accts'] = frame['num_tot_accts'].apply(lambda x : x if not pd.isnull(x) else means['num_tot_accts'])
frame['inq_last_6m'] = frame['inq_last_6m'].apply(lambda x : x if not pd.isnull(x) else means['inq_last_6m'])
frame['num_open_accts'] = frame['num_open_accts'].apply(lambda x : x if not pd.isnull(x) else means['num_open_accts'])
frame['revol_bal'] = frame['revol_bal'].apply(lambda x : x if not pd.isnull(x) else means['revol_bal'])
frame['accts_opn_last_6m'] = frame['accts_opn_last_6m'].apply(lambda x : x if not pd.isnull(x) else means['accts_opn_last_6m'])
frame['inq_last_12m'] = frame['inq_last_12m'].apply(lambda x : x if not pd.isnull(x) else means['inq_last_12m'])
frame['accts_opn_last_24m'] = frame['accts_opn_last_24m'].apply(lambda x : x if not pd.isnull(x) else means['accts_opn_last_24m'])
frame['earliest_cr_line'] = frame['earliest_cr_line'].apply(lambda x : x if not pd.isnull(x) else means['earliest_cr_line'])
frame['util_rate'] = frame['util_rate'].apply(lambda x : x if not pd.isnull(x) else means['util_rate'])

frame['dti'] = frame['dti'].apply(lambda x : x if not pd.isnull(x) else means['dti'])
frame['loan_over_income'] = frame['loan_over_income'].apply(lambda x : x if not pd.isnull(x) else means['loan_over_income'])

frame['home_ownership'] = frame['home_ownership'].apply(lambda x : x if not pd.isnull(x) else "NoData")
```

```python
frame['tot_credit_bal'] = frame['tot_credit_bal'].apply(lambda x : x if x > 0 else means['tot_credit_bal'])
frame['inq_last_6m'] = frame['inq_last_6m'].apply(lambda x : x if x > 0 else means['inq_last_6m'])
frame['num_open_accts'] = frame['num_open_accts'].apply(lambda x : x if x > 0 else means['num_open_accts'])
frame['revol_bal'] = frame['revol_bal'].apply(lambda x : x if x > 0 else means['revol_bal'])
frame['accts_opn_last_6m'] = frame['accts_opn_last_6m'].apply(lambda x : x if x > 0 else means['accts_opn_last_6m'])
frame['inq_last_12m'] = frame['inq_last_12m'].apply(lambda x : x if x > 0 else means['inq_last_12m'])
frame['accts_opn_last_24m'] = frame['accts_opn_last_24m'].apply(lambda x : x if x > 0 else means['accts_opn_last_24m'])
frame['earliest_cr_line'] = frame['earliest_cr_line'].apply(lambda x : x if x > 0 else means['earliest_cr_line'])
frame['util_rate'] = frame['util_rate'].apply(lambda x : x if x > 0 else means['util_rate'])
```

```python
frame.loc[(frame.platform_id == 'C'),'dti'] = frame['dti']-frame['monthly_payment']/(frame['income']+1)*12

frame['dti'] = frame['dti'].apply(lambda x : x if x > 0 else 0)

frame['post_dti'] =  frame['dti'] + frame['monthly_payment']/(frame['income']+1) * 12
```
