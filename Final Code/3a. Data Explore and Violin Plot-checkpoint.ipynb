```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
li = []
df_a_orig = pd.read_csv('Zach_Statham_platform_a_orig_data.csv', index_col=None, header=0)
li.append(df_a_orig)
df_b_orig = pd.read_csv('Zach_Statham_platform_b_orig_data.csv', index_col=None, header=0)
li.append(df_b_orig)
df_c_orig = pd.read_csv('Zach_Statham_platform_c_orig_data.csv', index_col=None, header=0)
li.append(df_c_orig)
```

```python
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
store = pd.HDFStore('df_all_orig.h5')
store['df_all_orig'] = frame  # save it
```

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.violinplot(x='platform_id', y='int_rate', hue="grade", hue_order= ["AA","A","B","C","D","E","F","G","HR"], data=frame, scale='width', inner='quartile')
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
plt.title("Violin Plot on interest rate across platforms, catergorized in grades")
plt.tight_layout()
```

```python
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.violinplot(x='platform_id', y='loan_amount', hue="grade", hue_order= ["AA","A","B","C","D","E","F","G","HR"], data=frame, scale='width', inner='quartile')
plt.legend(bbox_to_anchor=(1.01, 1),
           borderaxespad=0)
plt.title("Violin Plot on interest rate across platforms, catergorized in grades")
plt.tight_layout()
```

```python
frame['fico_category'] = pd.cut(frame.fico, [-np.inf, 650, 700, 750,800, np.inf],
                              labels=['<650','650-700', '700-750','750-800','>800'])
sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.violinplot(x='platform_id', y='loan_amount', hue="fico_category", hue_order= ['<650','650-700', '700-750','750-800','>800'], data=frame, scale='width', inner='quartile')
plt.legend(bbox_to_anchor=(1.5, 1),
           borderaxespad=0)
plt.title("Violin Plot on interest rate across platforms, catergorized in fico scores")
plt.tight_layout()
```
