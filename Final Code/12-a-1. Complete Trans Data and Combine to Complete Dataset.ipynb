```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_columns", None)
```

Sample Certain percent of data (e.g. 0.8 as 80%) and join with Orig data. 

```python
df_a_trans = pd.read_csv('data/Zach_Statham_platform_a_trans_data.csv', index_col=None, header=0)
df_a_trans_frac = df_a_trans

store = pd.HDFStore('df_all_orig.h5')
df_all_orig = store['df_all_orig']
df_a_all_frac = df_a_trans_frac.merge(df_all_orig, on='unique_id', how='left')
```

```python
df_a_trans[df_a_trans['default']==1].count()/df_a_trans.count()
```

```python
df_b_trans = pd.read_csv('data/Zach_Statham_platform_b_trans_data.csv', index_col=None, header=0)
df_b_trans_frac = df_b_trans

df_b_all_frac = df_b_trans_frac.merge(df_all_orig, on='unique_id', how='left')
```

```python
df_c_trans = pd.read_csv('data/Zach_Statham_platform_c_trans_data.csv', index_col=None, header=0)
df_c_trans_frac = df_c_trans

df_c_all_frac = df_c_trans_frac.merge(df_all_orig, on='unique_id', how='left')
```

```python
li = []
li.append(df_a_all_frac)
li.append(df_b_all_frac)
li.append(df_c_all_frac)
df_all_frac = pd.concat(li, axis=0, ignore_index=True)
```

```python
store = pd.HDFStore('df_all_frac.h5')
store['df_all_frac'] = df_all_frac  # save it
```
